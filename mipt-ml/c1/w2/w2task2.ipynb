```python
import math
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return math.sin(x/5.0) * math.exp(x/10.0) + 5.0 * math.exp(-x/2.0)

def F(X):
    return [f(x) for x in X]

def build_f_vector(x, n):
    return [x**i for i in np.arange(0, n)]

def build_M(X, n):
    return [build_f_vector(x, n) for x in X]

%matplotlib inline
def show_graph(W, n, start, end, num):
    x_plt = np.linspace(start, end, num)
    y_plt = np.dot(np.array([build_f_vector(x, n) for x in x_plt]), W)
    plt.plot(x_plt, y_plt)
    
    f_plt = F(x_plt)
    plt.plot(x_plt, f_plt)
    
    plt.show()
```

```python
n = 2
X = np.array([1., 15.])
W = np.linalg.solve(build_M(X, n), F(X))
show_graph(W, n, 0, 20, 10)
```

```python
n = 3
X = np.array([1., 8., 15.])
W = np.linalg.solve(build_M(X, n), F(X))
show_graph(W, n, 0, 20, 10)
```

```python
n = 4
X = np.array([1., 4., 10., 15.])
W = np.linalg.solve(build_M(X, n), F(X))
show_graph(W, n, 0, 20, 10)
```
